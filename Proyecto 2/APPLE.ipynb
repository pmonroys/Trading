### Relative Strength Index | RSI

The Relative Strength Index (RSI) is a popular technical analysis tool used by traders to assess the momentum of an asset and pinpoint potential shifts in its trend. By evaluating whether an asset is overbought or oversold, the RSI provides traders with valuable signals, enabling them to make more strategic decisions in the market. This concise overview highlights the key principles of the RSI strategy and its effectiveness in helping traders navigate financial markets with greater precision.


```python
strategy.plot_RSI()
```

In the lower graph, the Relative Strength Index (RSI) is depicted, which is a momentum-based indicator that oscillates between 0 and 100. It highlights whether the asset could be considered overbought (values above 70) or oversold (values below 30). When the RSI crosses these upper (overbought) or lower (oversold) boundaries, it may signal potential reversal points or trend continuations. However, traders often combine the RSI with other technical tools to enhance the accuracy of their trading decisions.

### Bollinger Bands
Bollinger Bands are a technical analysis tool consisting of three lines: a middle band (usually a moving average) and two outer bands that represent standard deviations from the middle. These bands expand and contract based on market volatility. When prices move near the upper band, the asset may be considered overbought, while prices near the lower band might indicate an oversold condition. Traders use Bollinger Bands to identify potential price breakouts, reversals, and to assess market volatility.


```python
strategy.plot_bollinger_bands()
```

The upper graph shows the closing prices of a financial asset, with Bollinger Bands overlaid to highlight price volatility. The middle line represents the asset‚Äôs moving average, while the outer bands indicate levels of standard deviation. When the price touches or moves close to the upper band, it may signal that the asset is overbought, while nearing the lower band could suggest an oversold condition. Traders often interpret price movements outside the bands as potential breakout signals, but it's essential to use Bollinger Bands alongside other indicators for more accurate trading decisions. The contraction or expansion of the bands also indicates changes in market volatility, with narrow bands suggesting low volatility and wide bands indicating high volatility.

### Williams %R
The Williams %R is a momentum oscillator that ranges from 0 to -100 and helps traders identify potential overbought or oversold conditions. In the strategy, a buy signal is triggered when the Williams %R falls below -80, indicating that the asset may be oversold, presenting a possible buying opportunity. A sell signal is generated when the Williams %R rises above -20, indicating the asset could be overbought and signaling a potential time to sell. This strategy aims to capture potential reversals in price trends by trading at these extreme conditions.

### MACD
The Moving Average Convergence Divergence (MACD) is a popular trend-following indicator that tracks the relationship between two moving averages of an asset's price. The strategy uses two main components: the MACD line, which is the difference between the short-term and long-term moving averages, and the Signal Line, which is the moving average of the MACD line. In this strategy, a buy signal is generated when the MACD line crosses above the Signal Line, suggesting upward momentum. A sell signal is triggered when the MACD line crosses below the Signal Line, indicating potential downward momentum.

### Stochastic Oscillator Strategy
The Stochastic Oscillator is a momentum indicator used to assess the position of an asset's price relative to its price range over a specific period. It consists of two lines: %K, which measures the current price relative to the high-low range, and %D, which is a moving average of %K. In this strategy, a buy signal is triggered when %K crosses above %D while both lines are below 20, indicating the asset may be oversold. Conversely, a sell signal is generated when %K crosses below %D while both lines are above 80, signaling that the asset could be overbought.

This strategy helps traders identify entry and exit points, but it is often used in combination with other technical indicators to improve accuracy and reliability in trading decisions.

### Ratios

**Sharpe Ratio**: The Sharpe Ratio measures the risk-adjusted performance of an investment. Specifically, it calculates the excess return of an investment compared to a risk-free rate of return, divided by the standard deviation of the investment returns, which represents the risk. A higher Sharpe Ratio indicates that an investment has had better risk-adjusted performance, as for every unit of risk, the investment has generated more excess return.

**Max Drawdown**: Max Drawdown represents the maximum loss that an investment could have experienced over a specific period. It is calculated as the largest drop from any peak to a subsequent low in the value of an investment account. This ratio is useful for measuring the maximum risk to which an investor was exposed and is considered an indicator of downward volatility.

**Win-Loss Ratio**: The Win-Loss Ratio compares the number of winning periods to losing periods in a series of investments. It is calculated by dividing the number of trades or periods that generated profits by the number of trades or periods that resulted in losses. This ratio helps to understand the consistency of the gains of an investment strategy and is commonly used to evaluate the effectiveness of a trader or a trading strategy. A ratio greater than 1 indicates that there were more winning periods than losing ones.
